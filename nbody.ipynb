```python
import numpy
import galpy
from galpy.orbit import Orbit
import matplotlib.pyplot as plt
```

```python
# Orbit mit Anfangspositionen und -geschwindigkeiten R, v_r, v_phi, z, v_z, phi
o = Orbit([1.,0.1,1.1,0.1,0.,0.])
# Potential: Plummer-Potential, entwickelt als Näherung für Kugelsternhaufen
# Phi = GM/sqrt(r^2+a^2) mit M=1.2, b=0.5
pot = galpy.potential.PlummerPotential(amp=1.2,b=0.5)
# Rechnung über 100 Zeiteinheiten, 10000 Zeitschritte
ts = numpy.linspace(0,100,10000)

o.integrate(ts,pot,method='dop853')

o.plot3d(xrange=(-1.2,1.2),yrange=(-1.2,1.2),zrange=(-1.2,1.2))
plt.show()
plt.plot(o.x(ts),o.y(ts))
plt.gca().set_aspect('equal')
plt.xlabel('x')
plt.ylabel('y')
#o.animate3d()
```

```python
# Orbit mit Anfangspositionen und -geschwindigkeiten R, v_r, v_phi, z, v_z, phi
o = Orbit([1.,0.1,1.1,0.1,0.,0.])
# Potential: Kuzmin-Potential für unendlich dünne Scheibe
# Phi = -GM/sqrt(R^2+(|z|+a)^2)
pot = galpy.potential.KuzminDiskPotential(amp = 1, a = 0.2)
# Rechnung über 100 Zeiteinheiten, 10000 Zeitschritte
ts = numpy.linspace(0,100,10000)
o.integrate(ts,pot,method='dop853')
o.animate3d()
```
